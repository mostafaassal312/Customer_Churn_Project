```python
# Goal : Understand why customers represent a churn risk.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load Data
df = pd.read_csv('../data/raw/telecom_churn_v1.csv')

# 2. Visual Style
sns.set_style("whitegrid")

# --- Insight 1: Does Contract Type Affect Churn? ---
plt.figure(figsize=(8, 5))
sns.countplot(x='Contract', hue='Churn', data=df, palette='Set1')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Number of Customers')
plt.show()
print("Observation: Customers with Month-to-month contracts are much more likely to churn.")

# --- Insight 2: Do Higher Charges Lead to Churn? ---
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='MonthlyCharges', hue='Churn', fill=True, palette='coolwarm')
plt.title('Monthly Charges Distribution: Churn vs No Churn')
plt.show()
print("Observation: Notice the peak? Customers paying high fees (> $90) are leaving faster.")

# --- Insight 3: Tenure vs Churn (Loyalty Check) ---
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='Tenure_Months', data=df, palette='pastel')
plt.title('Tenure (Months) vs Churn Status')
plt.show()
print("Observation: New customers (low tenure) are the most vulnerable to churning.")
```
